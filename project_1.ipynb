```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
```

```python
df = pd.read_csv("/content/BananaByte_Employee_Salary_090923 (1).csv")
```

```python
df['Perf'] = df['Perf'].fillna(df['Perf'].median())
```

```python
df.isnull().sum()
```

```python
df.drop(['Perf'],axis=1,inplace=True)
```

```python
df['Salary']=df['Salary'].astype('int')
```

```python
df['Salary'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Years_Exp':'Experience(yrs)'})
```

```python
df[['Salary','Bonus_FY', 'Total_Compensation']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Salary'].sum().sort_values(by='Salary',ascending=False)
sns.barplot(x='Gender',y='Salary',data=sales_gen)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Bonus_FY'].sum().sort_values(by='Bonus_FY',ascending=False)
sns.barplot(x='Gender',y='Bonus_FY',data=sales_gen)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Gender', y='Salary', palette='Set3', ci=None)
plt.title("Average Salary by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df, x='Edu_Level', y='Salary', order=df['Edu_Level'].value_counts().index, palette='Set1', ci=None)
plt.title("Average Salary by Education Level")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Job_Level', y='Salary', order=df['Job_Level'].value_counts().index, palette='viridis', ci=None)
plt.title("Average Salary by Job Level")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Years_Exp', y='Salary', hue='Gender', alpha=0.7)
plt.title("Experience vs Salary")
plt.show()
```
